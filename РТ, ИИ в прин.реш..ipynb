```python
import math

class Trig_func:
    
    def __init__(self, x):
        self.x = x
        
    def cos(self):
        return math.cos(self.x)
    def sin(self):
        return math.sin(self.x)
    def tan(self): 
        return math.tan(self.x)
    def arcsin(self):
        return math.asin(self.x)
    def arccos(self):
        return math.acos(self.x)
    def arctan(self):
        return math.atan(self.x)
    def rad(self):
        return math.radians(self.x)


func = int(input("   : \n"
                 "1: cos() \n"
                 "2: sin() \n"
                 "3: tan() \n"
                 "4: arcsin() \n"
                 "5: arccos() \n"
                 "6: arctan() \n"
                 "7: rad(    ) \n"))

x = int(input("  x: "))


obj = Trig_func(x)

if func == 1:
    print(obj.cos())
elif func == 2:
    print(obj.sin())
elif func == 3:
    print(obj.tan())
elif func == 4:
    print(obj.arcsin())
elif func == 5:
    print(obj.arccos())
elif func == 6:
    print(obj.arctan())
elif func == 7:
    print(obj.rad())
else:
    print(" ")
    
```

```python
class Tree:
    def __init__(self,data):
        self.left = None # 
        self.right = None # 
        self.data = data #
        
    def insert(self, value):
        if value < self.data:
            if self.left is None:
                self.left = Tree(value)
            else:
                self.left.insert(value)
        elif value > self.data:
            if self.right is None:
                self.right = Tree(value)
            else:
                self.right.insert(value)
        
    def Print_tree(self):
        tree_list = self._get_tree_list()
        for level, nodes in enumerate(tree_list):
            indent = "    " * level
            print(indent + " ".join(str(node) for node in nodes))

    def _get_tree_list(self):
        tree_list = []
        self._get_tree_list_helper(self, 0, tree_list)
        return tree_list

    def _get_tree_list_helper(self, node, level, tree_list):
        if node is None:
            return
        if level >= len(tree_list):
            tree_list.append([])
        tree_list[level].append(node.data)
        self._get_tree_list_helper(node.left, level + 1, tree_list)
        self._get_tree_list_helper(node.right, level + 1, tree_list)

tree = Tree(5)
tree.insert(3)
tree.insert(7)
tree.insert(2)
tree.insert(4)
tree.insert(6)
tree.insert(8)
tree.Print_tree()    
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree


X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

classifier = tree.DecisionTreeClassifier()
classifier = classifier.fit(X, target)
```

```python
#   
test_data = np.array([[0, 0], [4, 4]])
predictions = classifier.predict(test_data)

for data, prediction in zip(test_data, predictions):
    print(f": {data}, : {prediction}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = r'https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv'
dataset = pd.read_csv(url)
dataset.head()

print(dataset.shape)
dataset.describe()
```

```python
plt.scatter (dataset['fixed acidity'], dataset['quality'], color = 'b', label = "")
plt.xlabel("")
plt.ylabel("")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
print(X)
print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
from sklearn import tree
tree.plot_tree(regressor)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn import metrics 
print('MSE: ', metrics.mean_squared_error(y_test, y_pred))
print('MAE: ', metrics.mean_absolute_error(y_test, y_pred))
```

```python
metrics.mean_absolute_error(y_test, y_pred) / np.average(y) * 100
```
